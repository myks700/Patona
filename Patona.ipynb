```python
import selenium
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import warnings
warnings.filterwarnings('ignore')
import time
import requests
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(r'chromedriver.exe')
driver.get('https://dermnetnz.org/image-library')
```

```python
lists = []
image_url =[]

for page in range(1):
    name = driver.find_elements(By.XPATH, '//a[@class="imageList__group__item"]')
    image = driver.find_elements(By.XPATH, '//div[@class="imageList__group__item__image"]//img')
    
for i in range(len(name)):
    lists.append([name[i].text, name[i].get_attribute('href'), image[i].get_attribute('src')])
```

```python

for img in image:
    source = img.get_attribute('src')
    if source is not None:
         if (source[0:4] == 'http'):
            image_url.append(source)

for i in range(len(image_url)):
    if i > 294:
        breakBy.XPATH, 
    print("Downloading {0} of {1} images" .format(i, 294))
    response = requests.get(image_url[i])
    file = open(r'C:\Users\tajen\Pictures\Patona/'+str(i)+'.jpg','wb')
    file.write(response.content)
```

```python
DermNet = pd.DataFrame(lists, columns = ['Disease_Name', 'URL', 'Image_URL'])
DermNet.to_csv('DermNet.csv')
```
